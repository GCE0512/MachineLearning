```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100) # [0,14]的区间上找出100个点
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * 5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

```python
# 5种主题风格
# ·darkgrid
# ·whitegrid
# ·dark
# ·white
# ·ticks
```

```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
sns.violinplot(data)
sns.despine(offset=10) # 图标距离坐标轴的距离
```

```python
sns.set_style('whitegrid')
sns.boxplot(data=data, palette='deep')
sns.despine(left=True)
```

```python
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```
